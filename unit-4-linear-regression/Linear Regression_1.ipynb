# Linear Regression

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# 線性迴歸基本概念

直線

$f(x) = 1.2x + 0.8$

```python
# 準備 50 個點
# 5-0/50 = 0.1 +....
# 0~5線性切割
x = np.linspace(0, 5, 50)
print(x)
```

```python
y = 1.2*x + 0.8
y
```

```python
# 畫圖
plt.scatter(x, y)
plt.plot(x, 1.2*x + 0.8, 'r')
plt.show()
```

add noise

$f(x) + \varepsilon(x)$

```python
# 添加噪音
y = 1.2*x + 0.8 + 0.6*np.random.randn(50)
y
```

```python
# 添加噪音的圖
plt.scatter(x, y)
plt.plot(x, 1.2*x + 0.8, 'r')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
```

```python
regr
```

要求資料轉換<br>

$$[x_1, x_2, \ldots, x_{50}]$$

==> 

$$[[x_1], [x_2], \ldots, [x_{50}]]$$


矩陣的概念

`x.reshape(-1, 1)`的主要目的是將一維數組 `x` 轉換為二維數組 `x_reshaped`

這樣可以符合 sklearn 線性迴歸模型 LinearRegression 的輸入要求

sklearn 的大多數機器學習模型都要求輸入數據是二維數組，即`(n_samples, n_features)`的形狀

其中 n_samples 是樣本數量，n_features 是特徵數量

```python
print(x.shape)
x_reshaped = x.reshape(-1, 1)
print(x_reshaped.shape)
```

```python
regr.fit(x_reshaped, y)
```

```python
y_predict = regr.predict(x_reshaped)
y_predict
```

```python
# 訓練參數
regr.get_params()
```

```python
# 繪製原始數據和預測結果
plt.scatter(x, y, label='Original data')  # 原始數據
plt.plot(x, y_predict, 'ro', label='Fitted line')  # 預測數據

plt.xlabel('x')
plt.ylabel('y_predict')

plt.legend()
plt.show()
```

```python
# 計算均方誤差MSE
mse = np.sum((y_predict - y)**2) / len(y)
print("MSE:", mse)
```

```python
# Y=WX+b

# 獲取模型參數
W = regr.coef_  # 斜率
b = regr.intercept_  # 截距
print("Slope (W):", W)
print("Intercept (b):", b)
```

```python
# 將新的x資料帶入算出結果
# 使用模型進行預測
print("Prediction for x=3:", W * 3 + b)
print("Prediction for x=0.10204082:", W * 0.10204082 + b)
```
