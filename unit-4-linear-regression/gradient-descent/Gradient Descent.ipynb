# 梯度下降

梯度下降法的核心思想是通過沿著損失函數的梯度反方向更新參數，逐步減小損失函數的值

梯度下降法的定義是：

$w^{j+1}=w^j-\alpha\nabla E(w)，j 為迭代次數，\alpha 為學習率。$

其中 $\nabla$ 就是梯度，一次微分的意思。<br>
若是單純的 $y=ax+b$ 的情況，一次微分就是斜率。<br>
只是在多變量的情境中，斜率這個名稱太狹隘，於是在數學上就稱為「梯度」。<br>

梯度的算法就是對每個變量作一次偏微分。
例如，給一函數為：$f(x,y)=x^2+2xy+y^2$。<br>
                                 
計算其梯度為：$\mathrm{\nabla}\ f\left(x_1,x_2\right)=\left(\frac{\partial f\left(x_1,x_2\right)}{\partial x_1},\ \frac{\partial f\left(x_1,x_2\right)}{\partial x_2}\right)=\left(2x_1+1,\ 2x_2+1\right)$ <br>

當 $(x_1, x_2)=(1, 2)$ 時，$\mathrm{\nabla}\ f\left(1,\ 2\right)=\left(3,\ 5\right)$。
上述結果用白話文來說，即 $x_1$ 方向上的斜率為 $3$；$x_2$ 方向上的斜率為 $5$ 的意思。

回過頭來看剛剛給的函數：$E(w)=w^2$，
我們這就來「手動」跑一次梯度下降法。

假定，第一次隨機初始化點位，$w=5$。

### 1. 方向：將梯度的方向取負號，就是我們想要移動的方向。
### 2. 大小：由於學習率固定，因此梯度值愈大，每次移動的距離愈遠！

### 這個反覆迭代的過程會一直到終止條件出現為止，例如：
#### 1. 迭代次數達到某個值。
#### 2. 迭代後的 loss 值間的差異小於某個數。
#### 3. 程式執行總時間限制。

```python
import numpy as np
import matplotlib.pyplot as plt

# 目標函數:y=x^2
def func(x): return np.square(x)

# 目標函數一階導數:dy/dx=2*x
def dfunc(x): return 2 * x

def GD(x_start, df, epochs, lr):    
    """  梯度下降法。給定起始點與目標函數的一階導函數，求在epochs次反覆運算中x的更新值
        :param x_start: x的起始點    
        :param df: 目標函數的一階導函數    
        :param epochs: 反覆運算週期    
        :param lr: 學習率    
        :return: x在每次反覆運算後的位置（包括起始點），長度為epochs+1    
     """    
    xs = np.zeros(epochs+1)    
    x = x_start    
    xs[0] = x    
    for i in range(epochs):         
        dx = df(x)        
        # v表示x要改變的幅度        
        v = - dx * lr        
        x += v        
        xs[i+1] = x    
    return xs

# Main
# 起始權重
x_start = 10    
# 執行週期數
epochs = 20
# 學習率   
lr = 0.8   
# 梯度下降法 
x = GD(x_start, dfunc, epochs, lr=lr) 
print (x)
# 輸出：[-5.     -2.     -0.8    -0.32   -0.128  -0.0512]

color = 'r'    
#plt.plot(line_x, line_y, c='b')    
from numpy import arange
t = arange(-6.0, 6.0, 0.01)
plt.plot(t, func(t), c='b')
plt.plot(x, func(x), c=color, label='lr={}'.format(lr))    
plt.scatter(x, func(x), c=color, )    
plt.legend()

plt.show()
```

# 如何找出最佳參數使得LOSS最小

可以利用Gradient Descent方法找出最適參數$B_0$, $B_1$, $B_2$…,使得Loss最小

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# 生成一些隨機數據點
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# 初始化模型參數
m = 0
b = 0
learning_rate = 0.01
iterations = 50

# 梯度下降函數
def gradient_descent(X, y, m, b, learning_rate, iterations):
    N = len(y)
    history = []
    for i in range(iterations):
        y_pred = m * X + b
        D_m = (-2/N) * np.sum(X * (y - y_pred))
        D_b = (-2/N) * np.sum(y - y_pred)
        m -= learning_rate * D_m
        b -= learning_rate * D_b
        history.append((m, b))
    return history

# 獲取梯度下降過程中的參數歷史
history = gradient_descent(X, y, m, b, learning_rate, iterations)

# 創建動畫
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

def update(i):
    ax1.clear()
    ax2.clear()
    
    # 獲取當前迭代的 m 和 b
    m, b = history[i]
    
    # 繪製 ax1
    m_values = [0] + [m for m, b in history[:i+1]]
    b_values = [0] + [b for m, b in history[:i+1]]
    ax1.plot(b_values, m_values, color='green')  # 連接點的綠線
    ax1.scatter(b_values, m_values, color='black')  # 散點圖
    ax1.set_xlim(-1, 1)
    ax1.set_ylim(-0.5, 2)
    ax1.set_xlabel('line y-intercept (B)')
    ax1.set_ylabel('line slope (M)')
    ax1.set_title('Gradient Search')
    ax1.text(0.05, 1.5, f'iteration = {i+1}\nm = {m:.2f}\nb = {b:.2f}', fontsize=12)
    
    # 繪製 ax2
    ax2.scatter(X, y)
    ax2.plot(X, m * X + b, 'r')
    ax2.set_title('Points and Line')
    ax2.set_xlabel('X')
    ax2.set_ylabel('y')
    ax2.set_xlim(0, 2)
    ax2.set_ylim(0, 15)
    ax2.text(1, 13, f'y = {m:.2f}x + {b:.2f}', fontsize=12)

ani = animation.FuncAnimation(fig, update, frames=len(history), repeat=False)
ani.save("gradient_descent_animation.gif", writer="pillow")
plt.show()
```

# 該怎麼找到 w 跟 b? 對 w, b 微分

損失函數: $ L=(y-wx-b)^2 $

假設 x = -1, y = -3

$ L = (y-wx-b)^2 = (-3+w-b)^2 $

---------------------------------------------------
複合函數$f(g(x))$微分:

$ \frac{d}{dx} f(g(x)) = f'(g(x)) \cdot g'(x) $

---------------------------------------------------

對 𝑤 微分
$ \frac{dL}{dw} = \frac{d}{dw} \cdot (-3+w-b)^2 $
- 外層做一次微分: $ \frac{dL}{dw} = \frac{d}{dw} \cdot (-3+w-b)^2 = 2(-3+w-b) = -6+2w-2b $
- 內層做一次微分: $ \frac{dL}{dw} = (-3+w-b) = 1 $
  - 對𝑤微分，所以3,b是常數-，對常數微分為0，而變量w對自身的微分為 1
- 相乘: $(-6+2w-2b) \cdot 1 = -6+2w-2b $
- 所以 $ L'w = -6+2w-2b $

---------------------------------------------------

對 b 微分
$ \frac{dL}{db} = \frac{d}{dw} \cdot (-3+w-b)^2 $
- 外層做一次微分: $ \frac{dL}{dw} = \frac{d}{dw} \cdot (-3+w-b)^2 = 2(-3+w-b) = -6+2w-2b $
- 內層做一次微分: $ \frac{dL}{dw} = (-3+w-b) = 1 $
  - 對b微分，所以3,w是常數-，對常數微分為0，而變量-b對自身的微分為 -1
- 相乘: $(-6+2w-2b) \cdot -1 = 6-2w+2b $
- 所以 $ L'b = 6-2w+2b $

---------------------------------------------------
損失函數 𝐿 移動的方向跟梯度的方向相反
