# 皮爾遜相關係數

````python
import numpy as np
import matplotlib.pyplot as plt

# 設置隨機數生成種子，使結果可重現
np.random.seed(1)

# 生成50個在0到100之間的隨機整數
x = np.random.randint(0, 100, 50)

# 生成與x具有正相關並添加一些噪聲的數據
y = x + np.random.normal(10, 10, 50)

# 計算x和y之間的相關係數矩陣
coef = np.corrcoef(x, y)
print(coef)

# 繪製x和y的散點圖
plt.scatter(x, y)
plt.show()
````

````python
import numpy as np

x = np.array([8, 9, 10, 7, 8, 9, 5, 7, 9, 8])
y = np.array([12, 15, 16, 18, 6, 11, 3, 12, 11, 16])

# 計算 x 和 y 的均值
x_mean = np.mean(x)
y_mean = np.mean(y)

print("x_mean:", x_mean)
print("y_mean:", y_mean)

# 計算 x 和 y 的差
x_diff = x - x_mean
y_diff = y - y_mean

# 計算協方差
covariance = np.sum(x_diff * y_diff) / (len(x) - 1)

# 計算 x 和 y 的標準差
x_std = np.std(x, ddof=1)
y_std = np.std(y, ddof=1)

# 計算相關係數
correlation_coefficient = covariance / (x_std * y_std)

print("Covariance:", covariance)
print("x Standard Deviation:", x_std)
print("y Standard Deviation:", y_std)
print("Correlation Coefficient:", correlation_coefficient)
````

## 繪製消費者滿意度與再購買次數的散佈圖

````python
import numpy as np
import matplotlib.pyplot as plt

# 定義購買次數 (x) 和滿意度 (y)
x = np.array([8, 9, 10, 7, 8, 9, 5, 7, 9, 8])
y = np.array([12, 15, 16, 18, 6, 11, 3, 12, 11, 16])

# 計算購買次數和滿意度的均值
x_mean = np.mean(x)
y_mean = np.mean(y)

# 生成一條水平線，表示購買次數的均值
xpt1 = np.linspace(0, 12, 12)      # 在 x 軸上從 0 到 12 等間距生成 12 個點
ypt1 = [y_mean for xp in xpt1]     # 在 y 軸上對應這些 x 點的均值 y_mean

# 生成一條垂直線，表示滿意度的均值
ypt2 = np.linspace(0, 20, 20)      # 在 y 軸上從 0 到 20 等間距生成 20 個點
xpt2 = [x_mean for yp in ypt2]     # 在 x 軸上對應這些 y 點的均值 x_mean

# 繪製散點圖，顯示滿意度 vs 購買次數
plt.scatter(x, y)

# 繪製水平線，表示平均購買次數
plt.plot(xpt1, ypt1, 'g')

# 繪製垂直線，表示平均滿意度
plt.plot(xpt2, ypt2, 'b')

# 添加網格
plt.grid()

# 顯示圖表
plt.show()
````
